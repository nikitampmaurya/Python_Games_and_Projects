```python
#importing all the required libraries such as pandas, numpy, matplotlib and seaborn

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#reading the file  
df = pd.read_csv('Diwali Sales Data.csv',
                 encoding= 'unicode_escape')

df
```

```python
df.info()  #to get information about the dataframe
```

```python
#Data cleaning 
df.drop(['Status', 'unnamed1'], axis=1, inplace=True) #to drop the empty columns
```

```python
pd.isnull(df).sum(), df.shape #2. checking for null value
```

```python
# deleting 12 rows with null values
df.dropna(inplace=True) 
```

```python
df.shape #to check the no of rows and cols in dataframe after deleting 12 rows
```

```python
df.columns #to get names of all the columns
```

```python
# Changing the datatype of Marital Status into string
df['Marital_Status'] = df['Marital_Status'].astype('object') 
 # Changing the datatype of amount into integer
df['Amount'] = df['Amount'].astype('int64')
df.info()
```

```python
df
```

```python
df.loc[df['Marital_Status'] == 0, 'Marital_Status'] = "Yes"
df.loc[df['Marital_Status'] == 1, 'Marital_Status'] = "No"
```

```python
df
```

```python
df.columns #exploring the dataframe by retrieving all the columns in the dataframe
```

```python
df.rename(columns = {'Marital_Status':"Married"}, inplace=True)  #renaming martial status as married
```

```python
df
```

```python
df[["Age","Orders","Amount"]].describe() #To get the details statistics of customer's age, number of orders and amount spent
```

```python
A = sns.countplot(x = 'Gender', data = df, hue = 'Gender', order=['M','F'],alpha=0.7)
plt.title('Contribution of Buyers by Gender')

for x in A.containers:
    A.bar_label(x) 

#According to the information presented in the graph below, the primary buyers are female.
```

```python
B = sns.countplot(x='Age Group',data=df,hue='Gender',alpha = 0.7, palette = "Accent")
plt.title('Contribution of Buyers by Age Groups and Gender')

for x in B.containers:
    B.bar_label(x)

# majority of the costumers are between age 26-35 and within the age range, 
#Females are the leading population.
```

```python
#Graph for Total Amount vs Age Group
C = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
C
```

```python
sns.barplot(x = 'Age Group',y= 'Amount' ,data = C, hue = 'Age Group', alpha = 0.7)
plt.title('Amount contributed by different age groups')
plt.show()
```

```python
df
```

```python
#State vs Orders

D = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending = False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x = 'State', y = 'Orders', data = D, hue = 'State')
plt.title('Orders from different State')
plt.show()

"""
According to the below graph, the maximum number of orders were made in Uttar Pradesh followed by
Maharashtra and then Karnataka.
"""
```

```python
# State vs Amount
F = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending = False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x = 'State', y = 'Amount', data = F, hue = 'State')
plt.title('Amount from different State')
plt.show()
"""
According to the below graph, the highest amount was contributed by the state of Uttar Pradesh followed by
Maharashtra and then Karnataka.
"""
```

```python
E = sns.countplot(x='Married',hue='Married',palette = 'Accent',alpha = 0.7,data=df)
plt.title('No of Buyers by Marital Status')
for x in E.containers:
    E.bar_label(x)
plt.show()
#married people shop more than unmarried 
```

```python
F = df.groupby(['Married','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Married',y='Amount',data=F, hue='Gender',palette= 'Accent', alpha=0.7)
plt.title('No of Buyers by Marital Status and Gender')
plt.show()
#married women spent the most.
```

```python
#occupation count
sns.set(rc={'figure.figsize':(20,5)})
G = sns.countplot(data=df,x='Occupation', hue='Occupation', alpha=0.7)
plt.title('No of Buyers from different occupations')
for x in G.containers:
    G.bar_label(x)
plt.show()
#majority of the customers are from the IT sector and the healthcare industry 
```

```python
#amount vs occupation
G = df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=G, x='Occupation',y='Amount',hue='Occupation', alpha=0.7)
plt.title('Amount contributed from different occupations')
plt.show()
#Highest amount was invested by people from IT sector and healthcare industry
```

```python
sns.set(rc={'figure.figsize':(24,5)})
H = sns.countplot(data=df,x='Product_Category', hue='Product_Category')
plt.title('No of products ordered by category')
for x in H.containers:
    H.bar_label(x)
plt.show()
```

```python
J = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = J, x = 'Product_Category',y= 'Amount', hue = 'Product_Category')
plt.title('Amount contributed by different products category')
plt.show()
#Food was the most sold products
```

```python
I = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = I, x = 'Product_ID',y= 'Orders',hue='Product_ID')
plt.title('Demanding Product ID')
plt.show()

#gives the list of top 10 most sold product IDs
```
